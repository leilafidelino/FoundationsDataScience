### Medium complexity - Birthday Paradox

Birthday paradox concerns the probability that, in a set of `n` randomly chosen people, some pair of them will have the same birthday. You can read about this problem at https://en.wikipedia.org/wiki/Birthday_problem.

Write a function called **`has_duplicates()`**, that takes a list and returns `True` if there is any element that appears more than once. It should not modify the original list.

If there are 45 students in the class, what are the chances that any two students will have the same birthday? You can estimate this probability by generating random samples of 45 birthdays and checking for matches.

**Hints:**
1. you can generate random birthdays with the `randint` function in the `random` module. 
2. For simplicity, use the day number of the year, not the actual date
3. you can use the book solution as a starting point for this assignment: https://github.com/AllenDowney/ThinkPython2/blob/master/code/birthday.py or http://greenteapress.com/thinkpython2/code/birthday.py

The code should print out: 
- number of students
- number of iterations/samples
- list of duplicate days for each iteration, where duplicates are found
- probability

```python
from random import randint


def has_duplicates(list_bdays):
    '''(list of int) -> bool
    
    Takes a list and returns True if there is any element that appears more than once. It does not modify the original list.'''
    
    copy_list = list_bdays[:]
    copy_list.sort()

    same_day =[]
    duplicate = False
    
    for i in range(len(copy_list)-1):
        if copy_list[i] == copy_list[i+1]:
            same_day.append(copy_list[i])
            duplicate = True
    
    # Removes duplicates in the same_day list.
    global same_day_final
    same_day_final = []
    [same_day_final.append(day) for day in same_day if day not in same_day_final]
        
    return duplicate


def count_matches(num_students, num_iterations):
    ''' (int, int) -> int
    
    Generates a sample of birthdays and counts duplicates.

    num_students: how many students in the group
    num_iterations: how many groups to simulate '''
    
    count = 0
    global iterations_with_duplicates
    iterations_with_duplicates = {}
    
    for i in range(num_iterations) :
        list_bdays = [randint(1, 365) for i in range(num_students)]
        if has_duplicates(list_bdays):
            count += 1
            iterations_with_duplicates[i + 1] = same_day_final
            
            
    return count


def bday_paradox():
    """Runs the birthday simulation and prints the number of matches."""
    num_students = 45
    num_iterations = 1000
    count = count_matches(num_students, num_iterations)
    p = count/num_iterations

    print('Number of students:', num_students )
    print('Number of iterations:', num_iterations)
    print('Probability of at least one match:', p*100, '%')
    print('Iterations with duplicates and their corresponding duplicate day(s): \n', iterations_with_duplicates)
    
    
bday_paradox()
```

Given a set of 45 students and after 1000 iterations, the probability of at least one match is 93.8%. This value is in line with the expected theoretical probability given the same number of students which is 93.4%. See probability_calc below.

```python
#Compare results of bday_paradox with probability_calc.

def probability_calc(n):
    '''Returns a theoretical calculation of the probability that at least two people in a set of n number of people have the same birthday. '''
    exp = (n*(n-1))/2
    P = (1 - ((364/365)**exp)) * 100
    probability = print((round(P, 1)),'%')
    return probability
```
